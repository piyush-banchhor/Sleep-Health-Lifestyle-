```python
# Distribution of Sleep Duration
plt.figure()
plt.hist(df['Sleep Duration'], bins=15)
plt.title("Distribution of Sleep Duration")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Frequency")
plt.show()


#The distribution of sleep duration shows that most individuals sleep between 6 to 8 hours
```

```python
# Distribution of Stress Level
plt.figure()
plt.hist(df['Stress Level'], bins=10)
plt.title("Distribution of Stress Level")
plt.xlabel("Stress Level")
plt.ylabel("Frequency")
plt.show()

#Stress levels are mainly concentrated between moderate values.
```

```python
# BIVARIATE ANALYSIS
# -----------------------------

# Scatter plot: Sleep Duration vs Stress Level
plt.figure()
plt.scatter(df['Sleep Duration'], df['Stress Level'])
plt.title("Sleep Duration vs Stress Level")
plt.xlabel("Sleep Duration (hours)")
plt.ylabel("Stress Level")
plt.show()
```

```python
# CORRELATION
# -----------------------------

correlation = df['Sleep Duration'].corr(df['Stress Level'])
print(f"Correlation between Sleep Duration and Stress Level: {correlation:.3f}")
```

```python
#The scatter plot and correlation analysis indicate a negative relationship between sleep duration and stress level,
#  suggesting that individuals who sleep longer tend to experience lower stress.
#  This highlights the importance of adequate sleep in stress management and overall health.
```

```python
#      Question 3  Python Code (Group-Based Analysis) 
#  GROUP BY GENDER
# ---------------------------------

gender_group = df.groupby("Gender")[["Sleep Duration", "Stress Level"]].mean()

# Sleep Duration by Gender
plt.figure()
gender_group["Sleep Duration"].plot(kind="bar")
plt.title("Average Sleep Duration by Gender")
plt.xlabel("Gender")
plt.ylabel("Sleep Duration (hours)")
plt.show()

# Stress Level by Gender
plt.figure()
gender_group["Stress Level"].plot(kind="barh")
plt.title("Average Stress Level by Gender")
plt.xlabel("Gender")
plt.ylabel("Stress Level")
plt.show()
```

```python
# GROUP BY OCCUPATION
# ---------------------------------

occupation_group = df.groupby("Occupation")[["Sleep Duration", "Stress Level"]].mean()

# Sleep Duration by Occupation
plt.figure()
occupation_group["Sleep Duration"].plot(kind="bar")
plt.title("Average Sleep Duration by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Sleep Duration (hours)")
plt.xticks(rotation=90)
plt.show()

# Stress Level by Occupation
plt.figure()
occupation_group["Stress Level"].plot(kind="bar")
plt.title("Average Stress Level by Occupation")
plt.xlabel("Occupation")
plt.ylabel("Stress Level")
plt.xticks(rotation=90)
plt.show()
```

```python
#GROUP BY BMI CATEGORY

bmi_group = df.groupby("BMI Category")[["Sleep Duration", "Stress Level"]].mean()

# Sleep Duration by BMI Category
plt.figure()
bmi_group["Sleep Duration"].plot(kind="bar")
plt.title("Average Sleep Duration by BMI Category")
plt.xlabel("BMI Category")
plt.ylabel("Sleep Duration (hours)")
plt.show()

# Stress Level by BMI Category
plt.figure()
bmi_group["Stress Level"].plot(kind="bar")
plt.title("Average Stress Level by BMI Category")
plt.xlabel("BMI Category")
plt.ylabel("Stress Level")
plt.show()


```

```python
# DISPLAY GROUPED DATA
# ---------------------------------

print("\nAverage Sleep Duration & Stress Level by Gender:")
print(gender_group)

print("\nAverage Sleep Duration & Stress Level by Occupation:")
print(occupation_group)

print("\nAverage Sleep Duration & Stress Level by BMI Category:")
print(bmi_group)
```

```python
#The group-based analysis shows noticeable differences in sleep patterns across demographics. 
#Certain genders and occupations demonstrate longer average sleep duration and lower stress levels, indicating healthier sleep habits. 
#Individuals within the normal BMI category tend to sleep longer and report lower stress compared to overweight and obese groups.
#Overall, groups with higher sleep duration and lower stress levels can be considered to exhibit healthier sleep patterns.
```

```python
#   question 4
# SELECT RELEVANT COLUMNS
# ---------------------------------
features = [
    "Sleep Duration",
    "Physical Activity Level",
    "Stress Level",
    "Heart Rate"
]

data = df[features]
```

```python
# CALCULATE AVERAGES
# ---------------------------------

avg_sleep = df["Sleep Duration"].mean()
avg_stress = df["Stress Level"].mean()

print(f"Average Sleep Duration: {avg_sleep:.2f} hours")
print(f"Average Stress Level: {avg_stress:.2f}")
```

```python
# IDENTIFY INDIVIDUAL GROUPS
# ---------------------------------

# Individuals sleeping less than average
low_sleep = df[df["Sleep Duration"] < avg_sleep]

# Individuals with stress higher than average
high_stress = df[df["Stress Level"] > avg_stress]

# Individuals with BOTH low sleep and high stress
high_risk = df[
    (df["Sleep Duration"] < avg_sleep) &
    (df["Stress Level"] > avg_stress)
]


print("Number of individuals sleeping less than average:", len(low_sleep))
print("Number of individuals with higher than average stress:", len(high_stress))
print("Number of high-risk individuals (low sleep & high stress):", len(high_risk))
```

```python
# LIFESTYLE INSIGHTS
# ---------------------------------

print("\nAverage Lifestyle Metrics for High-Risk Individuals:\n")
print(high_risk[[
    "Physical Activity Level",
    "Heart Rate",
    "Daily Steps",
    "Quality of Sleep"
]].mean())
```

```python
#Statistical analysis shows that a significant portion of individuals sleep less than the average sleep duration while also experiencing above-average stress levels. 
#High-risk individuals tend to have lower physical activity levels, reduced sleep quality, and elevated heart rates.
#These findings suggest that inadequate sleep and high stress are closely linked with poor lifestyle habits. 
#Improving physical activity and stress management may contribute to better sleep health and overall well-being.
```
